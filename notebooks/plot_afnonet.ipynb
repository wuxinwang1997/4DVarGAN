## 绘制同化预报循环误差

```python
rmse_afnonet = xr.open_mfdataset(f"{FORECAST_DIR}/rmse_afnonet.nc", combine="by_coords")["z"].values

acc_afnonet = xr.open_mfdataset(f"{FORECAST_DIR}/acc_afnonet.nc", combine="by_coords")["z"].values

mae_afnonet = xr.open_mfdataset(f"{FORECAST_DIR}/mae_afnonet.nc", combine="by_coords")["z"].values
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 3.5))
axes[0].plot(np.arange(1, 29), rmse_afnonet[1:], label=f"FourCastNet")
axes[1].plot(np.arange(1, 29), acc_afnonet[1:], label=f"FourCastNet")
axes[2].plot(np.arange(1, 29), mae_afnonet[1:], label=f"FourCastNet")

for j in range(3):
    axes[j].set_xticks(np.arange(0, 32, 4))
    axes[j].set_xticklabels([0, 1, 2, 3, 4, 5, 6, 7])

axes[0].set_ylabel(f"RMSE [m$^2$ s$^{-2}$]")
axes[1].set_ylabel(f"ACC")
axes[2].set_ylabel(f"MAE [m$^2$ s$^{-2}$]")
lines, labels = fig.axes[0].get_legend_handles_labels()
# fig.legend(lines, labels, ncol=4, loc='lower center', bbox_to_anchor=(0.5, 0))
fig.text(0.5, 0, f"Lead Times [days]", ha='center')
plt.savefig(f"fourcastnet.pdf",dpi=300, bbox_inches="tight")
```
